```python
# DATA / ALGORITHM / MODULE 불러오기
# 위 !ls 했을 때 matching.py가 보이지 않는다면 아래 코드를 실행해주세요
%cd drive/MyDrive/Wello+Enterprise_Policy_service_5팀(신희택,윤다솜)
import pandas as pd
from keras.models import load_model
from matching import MATCHING

d_ent_final = pd.read_csv('DATA/d_ent_final.csv')
p_ent_final = pd.read_csv('DATA/new_p_ent_final.csv')
concat_rating = pd.read_csv('DATA/concat_rating.csv')
model = load_model('ALGORITHM/model1.h5')
```

# <MATCHING MODEL 사용법>
- 형식 : MATCHING(d_ent_final,p_ent_final,concat_rating,model, 업종(대분류), 기업규모,  지역(시도), 지역(시군구), 관심지원유형)
- 업종, 기업규모, 지역(시도), 지역(시군구) 값을 넣으면 추천 정책을 확인할 수 있음.

- 코드 분류
    * 업종(대분류): 아래 표 참조
    * 기업규모: 소상공인/중소기업/중견기업/대기업
    * 지역(시도): 전국, 서울특별시, 부산광역시, 대구광역시, 인천광역시, 광주광역시, 대전광역시, 울산광역시, 세종특별자치시, 경기도, 강원도, 충청북도, 충청남도, 전라북도, 전라남도, 경상북도, 경상남도, 제주특별자치도
    * 지역(시군구): 해당 지역(시도)에 맞는 시군구
    * 지원유형: 현금,현물/정보제공/복지/민원,상담,법률/봉사,기부/교육,용역/기타
    * 결측값일 경우 해당없음 값 입력

- 입력 예시)
    * 업종: 농업, 임업 및 어업
    * 기업규모: 중소기업
    * 지역(시도): 강원도
    * 지역(시군구): 춘천시
    * 지원유형: 현금,현물
- => MATCHING(d_ent_final,p_ent_final,concat_rating,model,'농업, 임업 및 어업','중소기업','강원도','춘천시','현금,현물')
